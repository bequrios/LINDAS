```python
import pandas as pd
import yaml
from pylindas.pycube import Cube

df = pd.read_csv('https://dam-api.bfs.admin.ch/hub/api/dam/assets/31646012/master')
display(df.head(20))
```

```python
df.GEO_NR = df.GEO_NR.apply(lambda x: "https://ld.admin.ch/country/CHE" if x == "CH" else f"https://ld.admin.ch/municipality/{x}")
```

```python
df = df.drop(columns=['GEO_NAME', 'CLASS_HAB', 'SOURCE', 'VALUE_PERIOD', 'UNIT_VALUE', 'OBS_CONFIDENCE', 'STATUS'])
```

```python
df.head(20)
```

```python
with open('description.yml', encoding="utf-8") as f:
    cube_yaml = yaml.safe_load(f)
```

```python
cube = Cube(df, cube_yaml, environment="TEST", local=True)
cube.prepare_data()
cube.write_cube()
cube.write_observations()
cube.write_shape()
cube.serialize("triples.ttl")
```
